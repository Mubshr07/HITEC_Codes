```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('processed_dataset.csv')
```

```python
sns.countplot(x="Stress_Level", data=df)
plt.title("Stress Level Distribution")
plt.show()
```

```python
sns.histplot(df['Study_Hours_Per_Day'], bins=10)
plt.title("Study Hours Distribution")
plt.show()
```

```python
corr = df.drop(['Student_ID'], axis=1).corr()  
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x="Stress_Level", y="Sleep_Hours_Per_Day", data=df)
plt.title("Sleep Hours vs Stress Level")
plt.show()
```

```python
sns.boxplot(x="Stress_Level", y="Study_Hours_Per_Day", data=df)
plt.title("Study Hours vs Stress Level")
plt.show()
```

```python
sns.boxplot(x="Stress_Level", y="GPA", data=df)
plt.title("GPA vs Stress Level")
plt.show()
```

```python
sns.pairplot(df.drop("Student_ID", axis=1), hue="Stress_Level")
plt.show()
```
